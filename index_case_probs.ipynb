```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
for N in [10, 20, 30, 40, 50]:
    xs = np.linspace(1, 5, 5)
    ys = np.power(1 - xs / N, N)
    plt.plot(xs, ys, label="N="+str(N))
plt.xlabel("Number of index cases")
plt.ylabel("Prob of no index cases ocurring")
plt.legend(loc="upper right")    
plt.show()
```
